```python
#使用Python中的requests和json庫，並且利用pandas庫中的json_normalize()函數將JSON格式的資料轉換為DataFrame格式的表格數據。
import requests
import json
import pandas as pd


#利用requests對API來源發送一個請求

facemask = ("https://raw.githubusercontent.com/kiang/pharmacies/master/json/points.json")
response = requests.get(facemask)

#請求回應的內容存成一個字串格式
d = response.text

# 將長得像 json 格式的字串解析成字典或列表
data = json.loads(d)

#使用pd.json_normalize()方法將JSON格式的資料轉換為DataFrame格式的表格數據。其中record_path='features'表示features是JSON資料中一個列表的名稱，並且該列表內的每個元素都會轉換為一個DataFrame的一行數據。
df = pd.json_normalize(data, record_path='features')
df
```

# 接下來，請你將取回來的資料利用 Python 語法計算各地區的藥局數量，完成後 Commit 成第二個版本：

```python
med_count = {}

# 填入欄位名稱
for d in data["features"]:
    conunty = d["properties"]["address"][0:3]
    if conunty not in med_count:
         med_count[conunty] = 0
    
    else:
      med_count[conunty] += 1

print(med_count)
```

```python
#各縣市總口罩數量(TotalMask)
#資料整理
df['properties.address'] = df['properties.address'].str.slice(0,3)
#資料整理
df['Total_Mask'] = df['properties.mask_adult'] + df['properties.mask_child']
#使用groupby和sum，過濾出我們要的資料
temp_data = df.loc[:,'Total_Mask'].groupby(df['properties.address']).sum('Total_Mask')
#因為上面的結果是series，為了方便後續應用所以再轉成DataFrame
commit2_df = pd.DataFrame({'City':temp_data.index, 'TotalMask':temp_data.values})
commit2_df
```

# 接下來請你計算出每個地區的成人剩餘口罩數量，並且將結果從大到小排列，完成後 Commit 第三個版本：

```python
#計算每個地區的成人剩餘口罩數量，並將結果從大到小排列

mask_count = {}
for d in data["features"]:
  county = d["properties"]["address"][0:3]
  adult = d["properties"]["mask_adult"]

  if county not in mask_count:
    mask_count[county] = adult
  else:
    mask_count[county] += adult

# 將結果從大到小排列
mask_count = dict(sorted(mask_count.items(), key=lambda item: item[1], reverse=True))
print(mask_count)
# {'台北市': 12345, '新北市': 45678 ...}
```

```python
使用data frame呈現數據資料(利用sort_values指定(by)排序，並由大到小排列(False)
#過濾出我們要的資料
temp_data3 = df.loc[:,'properties.mask_adult'].groupby(df['properties.address']).sum('properties.mask_adult')
#轉成DataFrame
commit3_df = pd.DataFrame({'City':temp_data3.index, 'Adult Mask': temp_data3.values})
commit3_df = commit3_df.sort_values(by="Adult Mask", ascending=False)
commit3_df
```
